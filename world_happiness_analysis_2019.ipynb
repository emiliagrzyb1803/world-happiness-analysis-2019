## Dataset Description

The project utilized the **World Happiness Report 2019** dataset, sourced from the Kaggle platform.
The data presents the level of happiness in individual countries worldwide in 2019.

The dataset contains information on various factors influencing quality of life, such as:
- general happiness index,
- GDP per capita,
- social support,
- life expectancy,
- level of personal freedom,
- generosity and perceptions of corruption.

This data allows for the analysis of the relationship between socioeconomic factors and happiness levels on a global scale.



## Importing the necessary libraries. Setting the chart style.

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

## Defining the class and loading the data.


```python
class HappinessAnalyzer:
    def __init__(self, filepath):

        self.data = pd.read_csv("2019.csv")
        print("Data loaded. Number of rows:", len(self.data))

    def show_data(self):
        return self.data.head()

    def ranking_top_countries(self, number=10):
        top = self.data.sort_values(by='Score', ascending=False).head(number)

        plt.figure(figsize=(10, 5))
        sns.barplot(x='Score', y='Country or region', data=top, palette='viridis')
        plt.title(f"Top {number} of the happiest countries")
        plt.xlabel("Score")
        plt.ylabel("Country")
        plt.show()

    def correlations(self):
        plt.figure(figsize=(10, 8))

        numeric_data = self.data.select_dtypes(include=['float64', 'int64'])

        sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
        plt.title("What influences happiness? (1 = strong influence, 0 = no influence)")
        plt.show()

    def money_and_happiness(self):
        plt.figure(figsize=(10, 6))
        sns.scatterplot(x='GDP per capita', y='Score', data=self.data, hue='Score', size='Score')
        plt.title("Relationship: Wealth (GDP) and Happiness")
        plt.xlabel("GDP per capita")
        plt.ylabel("Happiness Score")
        plt.show()
```

## Previewing the data.


```python
analysis = HappinessAnalyzer('2019.csv')
analysis.show_data()
```

## Displays the ranking of the 10 happiest countries.


```python
analysis.ranking_top_countries(10)
```

## Heat map and correlation analysis of factors influencing happiness levels.


```python
analysis.correlations()
analysis.money_and_happiness()
```

# Summary and conclusions from the analysis.

The analysis identified key factors influencing social well-being.

**1. Economic foundation**

There is a strong positive correlation between GDP per capita and happiness.
The scatter plot shows a clear linear trend – as a country's wealth increases, so does its score. This is also confirmed by the heat map, where the correlation coefficient for this pair of variables is one of the highest.

**2. The role of health and support**

Social security and physical health are almost as important as material factors.
The variables *Social support* and *Healthy life expectancy* demonstrate a very high correlation with the overall score. This means that aspects of quality of life and interpersonal relationships cannot be ignored when modeling happiness.

**3. The limited importance of generosity**

The level of generosity does not directly translate into a country's overall happiness score. On the heat map, this area is marked with a cool/neutral color, and the coefficient value is low. This suggests that generosity is a cultural trait independent of a country's overall well-being.

**4. Dominance of the Western Model**

Analysis of the top rankings indicates the dominance of countries that effectively combine a free market with extensive social policies (Scandinavian countries, Western Europe).

**General conclusion:**
Happiness, from a macro-social perspective, is the result of three pillars: **economic stability, public health, and social support**.
